```python
import numpy as np
import matplotlib.pyplot as plt

#Función para graficar figuras de lissajous en columna
def figuras_de_lissajous(N:int,A1:int,A2:int,ciclos:int):
    ciclo=ciclos*np.pi
    matriz=np.zeros((N,2))
    desfases=[0,np.pi/4,np.pi/2]
    W1=1
    W2=1
    control_des=0
    while control_des !=3:
        if W1 <= 5 and W2 <=5:
            for i in range(0,len(matriz)):
                theta=ciclo/(i+1)
                x=A1*np.sin(W1*theta)
                y=A2*np.sin(W2*theta + desfases[control_des])
                matriz[i,0]=x
                matriz[i,1]=y
                
           
            if W2 < 5:
                W2 += 1
            else:
                W1+=1
                W2=W1
        else:
            W1=1
            W2=1
            control_des +=1
        plt.scatter(matriz[:,0],matriz[:,1])
        plt.axis('off')
        #a = ptl.show()
        plt.show()
        #print(a)

print(figuras_de_lissajous(1000, 1, 1, 1000))
        
```
