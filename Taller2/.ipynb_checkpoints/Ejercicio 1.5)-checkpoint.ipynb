```python
import numpy as np
import matplotlib.pyplot as plt

#Se define un dominio por recorrer

x = np.linspace(0,2*np.pi)

#Se obtiene la discretización reflejada
h  = x[1]-x[0]
print(h)

#Se utiliza una función de ejemplo de comprobación (también se puede usar sympy)

cos = lambda x: np.cos(x)
f=cos

#Aquí se aplica la formula de la derivada cuarta

def FourthDerivative(x,f,h,a):
    array_values=[]
    for i in x:
        quadra=((f(i+2)-4*f(i+1)+6*f(i)-4*f(i-1)+f(i-2))/((h**4)))/a
        array_values.append(quadra)
    return array_values


#A partir de las gráficas se compara la estimación aproximada y la cuarta derivada
#según la fórmula.

plt.plot(x, np.cos(x))
plt.plot(x,FourthDerivative(x, f, h,3000))
```
