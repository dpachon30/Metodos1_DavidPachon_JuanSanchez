```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
A = np.array([[2,-1],[1,2],[1,1]])
y = np.array([2,1,4])

def GetFit(A,y,n):
    
    b = y 
    AT = np.dot(A.T,A)
    bT = A.T @ b

    xsol = np.linalg.solve(AT,bT)
    
    return xsol
    
print(GetFit(A,y,0))

x = np.linspace(-1,2,1000)

y_1 = 2*x - 2
y_2 = (1 -x)/2
y_3 = 4 - x

plt.plot(x,y_1)
plt.plot(x,y_2)
plt.plot(x,y_3)

```

```python
x = np.arange(-5,5,0.03)
y = np.arange(-5,5,0.03)

b = np.array([[2],[1],[4]])
X = np.zeros((2,1))
X[0]=1
mat = np.zeros((len(x),len(y)))

for i in range(len(x)):
  X[0] = x[i]
  for j in range(len(y)):
    X[1] = y[j]
    RESTO_1 = np.dot(A,X)
    R = RESTO_1 - b
    r =np.linalg.norm(R)
    mat[i][j] = r
    

X,Y = np.meshgrid(x,y)
#D = np.zeros((len(x),len(y)))


fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,mat,cmap='coolwarm')

index = np.where(mat == np.amin(mat))
print(index)
print(x[214])
print(y[181])
```
