```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
P2 = np.linspace(0.1,0.5,100)  #Par치metros de probabilidad
P1 = np.linspace(0.1,0.9,100) 
```

```python
p1,p2 = np.meshgrid(P1,P2)
```

```python
def ProbValues(a,b): #Funci칩n para el 치rbol
    return (a*b+2*(1-a)*b+2*a*(1-b)+(1-a)*(1-b))/4
```

```python
Z = ProbValues(p1,p2) #Distribuci칩n
```

```python
figura = plt.figure( figsize=( 15 , 15 ) )
ax1 = figura.add_subplot( 1 , 1 , 1 , projection='3d' )    
surf = ax1.plot_surface( p1 , p2 , Z , cmap=cm.coolwarm , antialiased=True )
figura.colorbar(surf, shrink=0.5)   
ax1.set_xlabel('P1',size=17)

plt.title('Probability췂s surface',size=20)
ax1.set_ylabel('P2',size=17)
ax1.view_init(45,85)  
ax1.scatter(pmin[0],pmin[1],minvalue,s=80,color='k')
ax1.text(pmin[0],pmin[1],minvalue, '%s' % ('Minumum Value'), size=12, zorder=1, color='k')
```
